```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('youth_smoking_drug_data_10000_rows_expanded.csv')
```

```python
df
```

```python
df.describe()
```

```python

year = df[df['Year'] == df['Year'].iloc[0]]
gender_counts = year['Gender'].value_counts()

# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%') 

plt.title("Gender Distribution for {}".format(df['Year'].iloc[0]))

# Show plot
plt.show()
```

```python
age = df['Age_Group'].head(12)
health = df['Mental_Health'].head(12)

# Figure Size
fig = plt.figure(figsize =(10, 7))

# Horizontal Bar Plot
plt.bar(age[0:10], health[0:10])

plt.xlabel("Age Group of Youth")
plt.ylabel("Mental Health of Youth")
plt.title("Mental Health of Various Age Groups")

# Show Plot
plt.show()
```

```python
peer = df['Peer_Influence'].head(12)
prevalence = df['Smoking_Prevalence'].head(12)

# Figure Size
fig = plt.figure(figsize =(10, 7))

plt.scatter(peer, prevalence, c ="green")

plt.xlabel("Peer Influence")
plt.ylabel("Smoking Prevalence")
plt.title("Relationship Between Peer Influence and Smoking Prevalence Among Youth")
# To show the plot
plt.show()
```
